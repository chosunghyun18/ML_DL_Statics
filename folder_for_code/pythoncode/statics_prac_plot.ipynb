```python
# 문제  3 번  > 정규 분포     u = 3 , 편차  2 
np.random.seed(0)
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 3)
plt.rcParams['font.size'] = 12
plt.rcParams['lines.linewidth'] = 5
mu1, sigma1 = 3.0, 2.0
x = np.linspace(-10, 20, 1000)
y1 = (1 / np.sqrt(2 * np.pi * sigma1**2)) * np.exp(-(x-mu1)**2 / (2 * sigma1**2))
plt.plot(x, y1, alpha=0.7, label=r'PDF of N(3.0, $2.0^2$)')
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.axis([-10, 20])
plt.show()
```

```python
# 6 주사위를 30 번  3 이 5 번 이상 나올 확률
def bin_p(k, n, p):
    nck = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    pd = nck * p**k * (1-p)**(n-k)
    return pd
result = 0
for k in range(5):
    result+=bin_p(k,30,1/6)
# "주사위를 30 번  3 이 0~4 번 나올 모든 확률":result

print("주사위를 30 번  3 이 5 번 이상 나올 확률:",1-result)
```

```python
#7 x ~ N (u =20 , sig = 3 )   p(x >= a)
     # 정규 분포(Normal Distribution)확률 밀도 함수
    
# p(x >= a) = 0.3   >> p(x<=a) =0.7  
inp =np.linspace(0,20,20000)
a = 0
for i in inp:
    p=st.norm.cdf([20+i],20,3)
    if p > 0.6999:
        a = 20+i
        #print(a)
        #print(p)
        break
print("a의 값:",round(a,4))
```

```python
#8  10sin(2pi x) 신호

#(a)   make signal
x = np.arange(0,3, 0.01)   # 1이상 2미만 구간에서 0.1 간격으로 실수 생성
np.around(x,3)
s=[]
for k in x:
    s.append(10*math.sin(2*np.pi*k))

#(b) u = 0 , std = sig len(s) = x 의개수    

sigs =[1,2,3,4,5]

def noisedraw(sig):
    n1=np.random.normal(0,sig,len(s))
    y1 =[]
    for k in range(len(s)):
        y1.append(x[k]+n1[k])

    line=plt.plot(x,y1)
    plt.setp(line, color='b', linewidth=1.0)
    plt.title("under sigma = "+ str(sig))
    plt.xlabel("x")
    plt.ylabel("s")
    plt.show()

for sigma in sigs:
    noisedraw(sigma)
    
```
